Visualizing the distribution of the target variable
Typically: 0 = No Heart Disease (Healthy), 1 = Heart Disease (Sick)


```python
plt.figure(figsize=(6, 4))
sns.countplot(data=raw_data, x='target', hue='target', palette='viridis', legend=False)
plt.title('Distribution of Target Variable (Healthy vs. Sick)')
plt.xlabel('Target (0 = Healthy, 1 = Sick)')
plt.ylabel('Count')
plt.show()
```

Generating a correlation matrix to identify linear relationships between features

```python
plt.figure(figsize=(12, 10))
correlation_matrix = raw_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Clinical Features')
plt.show()
```
